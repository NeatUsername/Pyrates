```python
import pymongo
import pandas as pd
import re

# Setup Mongo DB to store and later export array as JSON
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.pyrates_db
collection = db.items
```

```python
cursor = db["items"].find()

starting_df = pd.DataFrame(list(cursor))
```

```python
starting_df.head()
```

```python
def dms2dd(s):
    # example: s = """0°51'56.29"S"""
    # ['', 'E', '59', '°', '2', '’', '0']
    null1, direction, degrees, null2, minutes, null3, seconds = re.split(r"([A-Z°\’]+)", s)
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    if direction in ('S','W'):
        dd*= -1
    return dd

converted_df = pd.DataFrame(starting_df)

converted_df['Latitude DMS'] = converted_df['Latitude DMS'].apply(dms2dd)
converted_df['Longitude DMS'] = converted_df['Longitude DMS'].apply(dms2dd)

converted_df.head()
```

```python
converted_df["Decimal Minutes"] = str("[") + converted_df["Latitude DMS"].astype(str) +", "+ converted_df["Longitude DMS"].astype(str) + str("]")
converted_df.head()
```

```python
converted_reordered_df = converted_df[["_id", "url", "Incident Date", "Type of Attack", "Type of Vessel", "Decimal Minutes", "Latitude DMS","Longitude DMS", "Narrative"]]
converted_reordered_df.head()
```

```python
converted_reordered_df.to_json('data_split_format.json',orient='split',default_handler=str,index=False)
```
